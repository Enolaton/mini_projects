```python
import matplotlib.pyplot as plt
import pandas as pd
import FinanceDataReader as fdr
# HMM
df_hmm = fdr.DataReader('011200', '2021-01-31', '2025-11-20')
df_hmm = df_hmm.reset_index()  # Date  
# SCFI
df_scfi = pd.read_csv("data/SCFI_21_25.csv", encoding="utf-8-sig")
df_scfi = df_scfi.drop(columns=[" :  ($)"], errors="ignore")
df_scfi[""] = pd.to_datetime(df_scfi[""])
merged = pd.merge(
    df_hmm,
    df_scfi,
    left_on="Date",
    right_on="",
    how="inner"
)
# 
fig, ax1 = plt.subplots(figsize=(14, 7))
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)
ax1.plot(merged["Date"], merged["Close"], color='black')
ax1.set_ylabel("HMM ", color='black')
ax1.tick_params(axis='y', labelcolor='black')
ax1.grid(True)
ax2 = ax1.twinx()
ax2.plot(merged["Date"], merged[""], color='red')
ax2.set_ylabel("SCFI ", color='red')
ax2.tick_params(axis='y', labelcolor='red')
plt.tight_layout()
plt.show()
```

```python
corr_value = merged["Close"].corr(merged[""])
print("HMM  vs SCFI  :", np.around(corr_value,2)) #     
```
